```python
import numpy
import pandas as pd
```

```python
technologies = ({
     'Courses':["Spark","PySpark","Hadoop","Python","pandas"],
     'Fee' :[20000,25000,26000,22000,24000],
     'Duration':['30days','40days','35days','40days','60days'],
     'Discount':[1000,1500,2500,2100,2000]
               })
df = pd.DataFrame(technologies)
```

```python
df
```

```python
df['Period'] = df['Courses'].astype(str) + '-' + df['Duration']  # use '+' operator to combine wto columns
```

```python
df
```

```python
df['Period2'] = df[ ['Courses', 'Duration'] ].apply('-'.join, axis=1)  
# use apply() method to combine two string columns
```

```python
df
```

```python
# use agg() method to merge string columns
```

```python
df['Period3'] = df[['Courses', 'Duration']].agg('-'.join, axis=1)
```

```python
df
```

```python
# use Series.str.cat() to combine
df ['Period5'] = df['Courses'].str.cat(df['Duration'], sep='-')
```

```python
df
```

```python
# use Apply() and lambda function
df['Period6'] = df[['Courses', 'Duration']].apply(lambda x:"==".join(x), axis=1)
```
